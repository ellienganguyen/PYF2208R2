```python
start = int(input("Input the start number: "))
end = int(input("Input the end number: "))

if start > end:
    print(f"The end number should be more than the start number")
else:
    for i in range(start, end+1): 
        if i % 3 == 0:
            if i % 5 == 0:
                print("FizzBuzz")
            else:
                print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)
        
```
