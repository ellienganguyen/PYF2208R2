Vẽ đường xoắn ốc như hình dưới sử dụng vòng lặp while True. Thoát khỏi vòng lặp khi khoảng cách từ điểm hiện tại đến điểm ban đầu lớn hơn một giá trị được nhập vào từ người dùng.


```python
import turtle
import random

t = turtle.Turtle()
colors = ['red', 'green', 'yellow', 'black', 'blue', 'orange', 'pink', 'brown']
d = 0.5
while d < 25:
    color = random.choice(colors)
    t.forward(d)
    t.right(10)
    d = d + 0.1
    t.color(color)

turtle.done()
```
