```python
import turtle

t = turtle.Turtle()
#t.hideturtle()
t.pencolor('brown')

def rectangle(w, h):
    for i in range(1,3):
        t.forward(w)
        t.right(90)
        t.forward(h)
        t.right(90)
    
def window(wd, hd):
    # Xác định vị trí vẽ cửa sổ
    t.penup()
    t.forward(wd)
    t.right(90)
    t.forward(hd)
    t.left(90)
    t.pendown()
    # Vẽ cửa sổ
    t.fillcolor('light blue')
    t.begin_fill()
    rectangle(wd, hd)
    t.end_fill()
    
def house():
    t.fillcolor('yellow')
    t.begin_fill()
    rectangle(w, h)
    t.end_fill()
    wd = w / 3
    hd = h / 3
    window(wd, hd)
    # Trở về vị trí ban đầu
    t.penup()
    t.goto(x, y)
    t.pendown()
    
def floor(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()
    house()
    t.penup()
    t.goto(x, y)
    t.forward(w)
    t.pendown()
    house()
      
w = 100 #float(input("Input the width: "))
h = 150 #float(input("Input the height: "))  
x = 0
y = 0
x1 = x
y1 = y
floor(x1, y1) # Vẽ tàng 1
# Di chuyển tới điểm vẽ tầng 2
x2 = x
y2 = y + 150
floor(x2, y2) # Vẽ tầng 2
#Di chuyển tới điểm vẽ Gara
t.penup()
t.goto(x=x+w*2, y=0)
t.pendown()
wg = h
hg = h
#Vẽ Gara
t.fillcolor('yellow')
t.begin_fill()
rectangle(wg, hg) 
t.end_fill()
#Di chuyển tới điểm vẽ cửa Gara
t.penup()
t.goto(x=2*w+h/4, y= -h/2)
t.pendown()
t.fillcolor('blue')
t.begin_fill()
wwg = h/4
hwg = h/2
rectangle(wwg, hwg)
t.penup()
t.goto(x=2*w+h/2, y= -h/2)
t.pendown()
rectangle(wwg, hwg)
t.end_fill()

turtle.done()
```
