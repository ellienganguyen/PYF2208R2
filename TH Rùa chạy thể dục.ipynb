Viết chương trình để rùa chạy từ bên trái màn hình sang bên phải, lúc rùa chạy các nét vẽ không liền nhau, giá trị độ dài các nét vẽ tùy ý.

```python
import turtle
import random as r

# khởi tạo turtle
t = turtle.Turtle()
# vẽ vạch xuất phát
t.pensize(8)
t.color("blue")
t.speed(5)
t.penup()

t.goto(-320, 150)
t.pendown()
t.right(90)
t.forward(300)
t.left(90)
t.penup()

# vẽ vạch đích
t.goto(280, 150)
t.pendown()
t.right(90)
t.forward(300)
t.left(90)

# Viết chữ
t.penup()
t.goto(-300, 200)
t.pendown()
t.color('purple')
style = ('Courier', 20, 'normal')
t.write('START', font=style, align='center')

t.penup()
t.goto(300, 200)
t.pendown()
t.color('purple')
style = ('Courier', 20, 'normal')
t.write('GOAL', font=style, align='center')
t.hideturtle()

t.pensize(3)
t.color("green")
t.speed(1)
t.penup()

t.shape("turtle")
# Ẩn hình ảnh rùa
t.hideturtle()
# Đặt vị trí ban đầu của con rùa sang bên trái so với vị trí giữa màn hình 400 pixel
# Mục đích để rùa chạy không ra khỏi màn hình khi vòng lặp quá lớn
t.goto(-320, 0)
# Hiển thị hình ảnh con rùa
t.showturtle()

count = 0
while count < 15:
    # sinh hai giá trị ngẫu nhiên
    down = r.randint(15, 25)
    up = r.randint(15, 25)
    t.pendown()
    # rùa tiến về phía trước với giá trị ngẫu
    # nhiên ở trên, có để lại nét vẽ
    t.forward(down)
    t.penup()
    # rùa tiến về phía trước với giá trị ngẫu
    # nhiên ở trên, không để lại nét vẽ
    t.forward(up)
    count += 1
#Viết chữ
t.penup()
t.goto(0, -200)
t.pendown()
t.color('Red')
style = ('Arial', 20, 'normal')
t.write('CONGRATULATION!!!', font=style, align='center')

turtle.done()
```
